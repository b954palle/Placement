```python
import pandas as pd 

dt = '20110102'
dt = pd.to_datetime(dt)
```

```python
# get time fields
dt
```

```python
print (dt.year)

print (dt.month)

print (dt.day)

print (dt.hour)

print (dt.minute)

print (dt.second)
```

```python
# handles hours/minute/seconds too
pd.to_datetime('20110102 12:10:30')
```
